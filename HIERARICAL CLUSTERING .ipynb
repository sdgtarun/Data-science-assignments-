### 1.	Perform clustering for the airlines data to obtain optimum number of clusters. Draw the inferences from the clusters obtained. Refer to EastWestAirlines.xlsx dataset.

```python
#import needed packages 
import pandas as pd
import matplotlib.pylab as plt 
```

```python
#this reads the file and loads sheet 2. 
df=pd.read_excel("C:\\Users\\tarun\\Documents\\taruns imp\\Data science material\\assignments\\SETS\\HIerarical clustering sets\\EastWestAirlines.xlsx",sheet_name='data')
```

```python
#drops unnessary columns 

df_1=df.drop(['ID#','Award?'],axis=1)
```

```python
# Normalization function 
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x) 
```

```python
# Normalized data frame 
df_norm=norm_func(df_1.iloc()[:,1:])
df_1.describe
```

```python
# for creating dendrogram 
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
z=linkage(df_norm,method='complete',metric='euclidean')
```

```python
#dendrogram
plt.figure(figsize=(100,70))
plt.title("dendrogram")
sch.dendrogram(z,
                leaf_rotation=0,
               leaf_font_size=10)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_complete = AgglomerativeClustering(n_clusters = 3, linkage = 'complete', affinity = "euclidean").fit(df_norm) 
h_complete.labels_
```

```python
cluster_labels = pd.Series(h_complete.labels_)
```

### 3.	Perform clustering analysis on the telecom data set. The data is a mixture of both categorical and numerical data. It consists of the number of customers who churn out.
```python
#load necessary packages 
import pandas as pd
import matplotlib.pylab as plt 
```

```python

df=pd.read_excel("C:\\Users\\tarun\\Documents\\taruns imp\\Data science material\\assignments\\SETS\\hierarical clustering sets\\Telco_customer_churn.xlsx")
```

```python
#selectes only neede colums 
df2=df[['Tenure in Months','Avg Monthly Long Distance Charges','Avg Monthly GB Download','Monthly Charge','Total Charges','Total Refunds','Total Extra Data Charges','Total Long Distance Charges','Total Revenue']]
```

```python
df2
```

```python
#Normalization function 
def norm_func(i):
    x = (i-i.min())	/ (i.max()-i.min())
    return (x)
```

```python
#normatized data
norm_df=norm_func(df2.iloc()[:,1:])
```

```python
#dendrogram
plt.figure(figsize=(15,8))
plt.title("dendrogram")
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
cluster.fit_predict(norm_df)
```
