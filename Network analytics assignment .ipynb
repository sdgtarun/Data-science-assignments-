```python
connt.head()
```

```python
fg = nx.from_pandas_edgelist(connt, source = "source_airport", target = "Destination ID",edge_attr=True)
```

```python
print(nx.info(fg))
```

```python
fg.nodes()
```

```python
fg.edges
```

```python
b=nx.degree_centrality(fg)
print(b)
```

```python
pos=nx.spring_layout(fg,k=0.15)
```

```python
nx.draw_networkx(fg, pos, node_size = 25, node_color = 'blue')
```

```python
closeness =nx.closeness_centrality(fg)
```
